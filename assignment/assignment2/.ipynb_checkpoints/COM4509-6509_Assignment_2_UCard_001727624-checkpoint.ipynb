```python
# Write the code for your answer here. You can use multiple cells to improve readability.



import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

## 1a. Download dataset
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=1, shuffle=False, num_workers=2)

testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=1, shuffle=False, num_workers=2)

classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
#1b Add random noise to all training and test data to generate noisy dataset
card_id = 1727624
torch.manual_seed(card_id)

def random_noise(loader, scale=0.2):
    noise = []
    for data in loader:
        img, label = data[0], data[1]
        img_noise = img + scale * torch.randn_like(img)
        noise.append(img_noise.numpy())
    noise_data = np.array(noise)
    return np.clip(noise_data, -1, 1)

noise_train_data = random_noise(trainloader)
noise_test_data = random_noise(testloader)
print("add noise")
```

```python
#1c Extract a subset with only two classes: Cat and Dog and name it starting with CatDog.

#getting noisy train and test tensors
noise_train_tensor = torch.from_numpy(noise_train_data).squeeze()
noise_test_tensor = torch.from_numpy(noise_test_data).squeeze()

def get_subclass(loader, noise):
    sub_class = []
    sub_class_noise = []
    sub_class_label = []
    for index, data in enumerate(loader):
        img, label = data[0], data[1]
        if label == 3 or label == 5:
            sub_class.append(img.numpy())
            sub_class_label.append(label.numpy())
            sub_class_noise.append(noise[index, :, :, :].numpy())
    return torch.from_numpy(np.array(sub_class)).squeeze(), torch.from_numpy(np.array(sub_class_label)).squeeze(), torch.from_numpy(np.array(sub_class_noise)).squeeze()
        
CatDog_trainset, CatDog_trainlabel, CatDog_trainnoise = get_subclass(trainloader, noise_train_tensor)
CatDog_testset, CatDog_testlabel, CatDog_testnoise = get_subclass(testloader, noise_test_tensor)
```

```python
#1d
#extract dog and cat index
cat_index = (torch.eq(CatDog_trainlabel, 3)).nonzero()
dog_index = (torch.eq(CatDog_trainlabel, 5)).nonzero()

show_cats = CatDog_trainset[cat_index[:10], :, :, :].squeeze()
show_cats_noise = CatDog_trainnoise[cat_index[:10], :, :, :].squeeze()

show_dogs = CatDog_trainset[dog_index[:10], :, :, :].squeeze()
show_dogs_noise = CatDog_trainnoise[dog_index[:10], :, :, :].squeeze()

def imshow(img_cat, img_cat_noise, img_dog, img_dog_noise):
    plt.subplot(411)
    img_cat = img_cat / 2 + 0.5
    npimg_cat = img_cat.numpy()
    plt.imshow(np.transpose(npimg_cat, (1, 2, 0)))
    
    plt.subplot(412)
    img_cat_noise = img_cat_noise / 2 + 0.5
    npimg_cat_noise = img_cat_noise.numpy()
    plt.imshow(np.transpose(npimg_cat_noise, (1, 2, 0)))
    
    plt.subplot(413)
    img_dog = img_dog / 2 + 0.5
    npimg_dog = img_dog.numpy()
    plt.imshow(np.transpose(npimg_dog, (1, 2, 0)))
    
    plt.subplot(414)
    img_dog_noise = img_dog_noise / 2 + 0.5
    npimg_dog_noise = img_dog_noise.numpy()
    plt.imshow(np.transpose(npimg_dog_noise, (1, 2, 0)))
    
    plt.show()

imshow(torchvision.utils.make_grid(show_cats, nrow=10), 
    torchvision.utils.make_grid(show_cats_noise, nrow=10),
      torchvision.utils.make_grid(show_dogs, nrow=10), 
       torchvision.utils.make_grid(show_dogs_noise, nrow=10))
```

```python
# 2a Apply PCA on the training set to reduce the dimensionality.
from sklearn.decomposition import PCA

b_train, _, _, _ = CatDog_trainset.size()
train_X = CatDog_trainset.view(b_train, -1).numpy()
train_y = CatDog_trainlabel.numpy()
b_test, _, _, _ = CatDog_testset.size()
test_X = CatDog_testset.view(b_test, -1).numpy()
test_y = CatDog_testlabel.numpy()


k = [3, 4, 8, 14, 32, 100, 500]
pca_list = []
pca_test_list = []
for k_index in k:
    estimator = PCA(n_components=k_index)
    pca_x = estimator.fit_transform(train_X)
    pca_list.append(pca_x)
    
    pca_test_x = estimator.fit_transform(test_X)
    pca_test_list.append(pca_test_x)
    print("k:",k_index)
    print(sum(estimator.explained_variance_ratio_))
    
# 500, 0.9913; 100 0.9166; 32 0.8071; 14 0.7012; 8 0.6088; 4 0.5; 3, 0.4407
```

The reason for choosing k as 3,4,8,14,32,100,500 is every k is corresponding to a raito of variance. The ratio of variance stands for how much data in keeping original feature.

```python
#2b Train eight Naive Bayes Classifiers (NBC)

#Seven naive bays classifiers on seven PCA features.
from sklearn.naive_bayes import GaussianNB
gnb_list = []
for i in range(len(pca_list)):
    gnb = GaussianNB()
    gnb.fit(pca_list[i], train_y)
    gnb_list.append(gnb)
    
#One naive bays classifiers on original
gnb = GaussianNB()
gnb.fit(train_X, train_y)
gnb_list.append(gnb)

```

```python
# 2c Evalaute the eight Naive Bayes classifiers
accuracy_list = []
predicts = []
for i in range(len(gnb_list)):
#     print(pca_test_list[i].shape)
    if i == len(gnb_list) - 1:
        y_predict = gnb_list[i].predict(test_X)
    else:
        y_predict = gnb_list[i].predict(pca_test_list[i])
    predicts.append(y_predict)
    accuracy = sum(y_predict == test_y) / len(test_y)
    accuracy_list.append(accuracy)
print(accuracy_list)
x = np.arange(8)
plt.bar(np.array(x), height=np.array(accuracy_list), width=0.8)
plt.show()
```

```python
# 2d Plot the ROC Curves in true positive rates vs false positive rates for the eight Naive Bayes classifiers in one figure

from sklearn.metrics import roc_curve, auc
roc_label = k.copy()
roc_label.append(32*32)
roc_auc = {}
for i in range(len(predicts)):
    fpr, tpr, thresholds = roc_curve(predicts[i]==5, test_y==5)
    roc_auc[i] = auc(fpr, tpr)
    plt.plot(fpr, tpr, label="PCA_dim_" + str(roc_label[i]))
plt.legend()
plt.show()
```

The reason for choosing Gaussian model is because gaussian model assumes the feature value observed is in gaussian distribution. It is common assumption we make typically in image processing.

```python
# 2e Compute the area under the ROC curve values for the eight Naive Bayes classifiers and visualise

roc_area_y = np.array([item for key, item in roc_auc.items()])
plt.bar(x, roc_area_y)
plt.show()
```

The training with raw data has third best result in accuracy. The best accuracy result is come from a Naive Bays on 500 PCA features.

```python
# 3b Train nine classifers


train_data = torch.cat([data[0] for data in trainloader], dim=0)
test_data = torch.cat([data[0] for data in testloader], dim=0)
print(train_data.size())
```

```python
train_all_X = train_data.view(50000, -1).numpy()
test_all_X = test_data.view(10000, -1).numpy()

# 4 0.50; 20 0.74; 400 0.97
k = [4, 20, 400]
pca_all_list = []
pca_all_test_list = []
for k_index in k:
    estimator = PCA(n_components=k_index)
    pca_all_x = estimator.fit_transform(train_all_X)
    pca_all_list.append(pca_all_x)
    print(sum(estimator.explained_variance_ratio_))
    
    pca_all_test_x = estimator.fit_transform(test_all_X)
    pca_all_test_list.append(pca_all_test_x)
    print(sum(estimator.explained_variance_ratio_))
```

```python
train_all_y = torch.cat([data[1] for data in trainloader], dim=0).numpy()
test_all_y = torch.cat([data[1] for data in testloader], dim=0).numpy()
```

```python
# training Four NB models

import time
classifier_list = []
training_time = []
for i in range(3):
    gnb = GaussianNB()
    start = time.time()
    gnb.fit(pca_all_list[i], train_all_y)
    end = time.time()
    cost = end - start
    classifier_list.append(gnb)
    training_time.append(cost)
gnb = GaussianNB()
start = time.time()
gnb.fit(train_all_X, train_all_y)
end = time.time()
cost = end - start
training_time.append(cost)
classifier_list.append(gnb)
print("success NB")
```

```python
# Four LR models

from sklearn.linear_model import LogisticRegression
for i in range(3):
    lr = LogisticRegression(max_iter=50, tol=0.01,n_jobs=-1)
    print("start training")
    start = time.time()
    lr.fit(pca_all_list[i], train_all_y)
    end = time.time()
    cost = end - start
    print("end training")
    classifier_list.append(lr)
    training_time.append(cost)
lr = LogisticRegression(max_iter=50, tol=0.01,n_jobs=-1)
print("start training")
start = time.time()
lr.fit(train_all_X, train_all_y)
end = time.time()
cost = end - start
print("end training")
training_time.append(cost)
classifier_list.append(lr)
print("success LR")
```

```python
#Training CNN

import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 16 * 5 * 5)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x


net = Net()
print("construct net")
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
trainloadernet = torch.utils.data.DataLoader(trainset, batch_size=16, shuffle=True, num_workers=2)
print("start training")
start = time.time()
for epoch in range(10):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(trainloadernet, 0):
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 2000 == 1999:    # print every 2000 mini-batches
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, running_loss / 2000))
            running_loss = 0.0
end = time.time()
cost = end - start
training_time.append(cost)
print('Finished Training')
PATH = './cifar_net.pth'
torch.save(net.state_dict(), PATH)
```

```python
# Write the code for your answer here. You can use multiple cells to improve readability.

# 4a Modify the autoencoder architecture so that it takes colour images as input

class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            # 1 input image channel, 16 output channel, 3x3 square convolution
            nn.Conv2d(3, 16, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 32, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(32, 64, 7)
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(64, 32, 7),
            nn.ReLU(),
            nn.ConvTranspose2d(32, 16, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(16, 3, 3, stride=2, padding=1, output_padding=1),
            nn.Sigmoid()  #to range [0, 1]
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
# 4b Training: feed the noisy training images as input to the autoencoder in 4a; use a loss function that computes the reconstruction error between the output of the autoencoder and the respective original images.

net_train_data = torch.utils.data.TensorDataset(noise_train_tensor, train_data)
net_test_data = torch.utils.data.TensorDataset(noise_test_tensor, test_data)
net_train_loader = torch.utils.data.DataLoader(net_train_data, batch_size=16, shuffle=True, num_workers=2)
net_test_loader = torch.utils.data.DataLoader(net_test_data, batch_size=1, shuffle=False, num_workers=2)
net = Autoencoder()
criterion = nn.L1Loss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
trainloadernet = torch.utils.data.DataLoader(trainset, batch_size=16, shuffle=True, num_workers=2)
print("start training")
start = time.time()
for epoch in range(10):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(net_train_loader, 0):
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 2000 == 1999:    # print every 2000 mini-batches
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, running_loss / 2000))
            running_loss = 0.0
print("finish training")
torch.save(net.state_dict(), './cifar_denoise.pth')
```
